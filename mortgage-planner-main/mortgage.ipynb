```python

calculator = financialEstimator( 
                rent_price_growth = RENT_GROWTH,
                rent_price = RENT_PRICE,
                house_price_growth = HOUSE_PRICE_GROWTH,
                house_price = HOUSE_PRICE,
                monthly_mortgage_interes = INTEREST,
                intallment_threshold = INSTALLMENT_THRESH,
                savings_per_month = SAVINGS_PER_MONTH,
                etf_growth = ETF_GROWTH,
                initial_etf_savings = 0 # INITIAL_ETF_SAVINGS
                )
```

```python
opt_result, opt_verbose = calculator.run_simulation()

# Print the results
print("Optimal solution:", opt_result.x)
print("Optimal value of the objective function:", opt_result.fun)


import pandas as pd

sub_opt_steps = pd.DataFrame(opt_verbose.decreasing_list_calls_inp, columns = ['months_to_wait', 'mortgage_years'])
sub_opt_steps['cost_function'] = opt_verbose.decreasing_list_calls_res
```
